geeksofgeeks question : https://www.geeksforgeeks.org/problems/find-minimum-and-maximum-element-in-an-array4428/1


```python
class Solution:
    def getMinMax(self, arr):

        def solve(arr, strt, end):
            # base case
            if strt == end:
                return arr[strt], arr[strt]

            minn, maxx = solve(arr, strt + 1, end)

            if arr[strt] < minn:
                minn = arr[strt]
            if arr[strt] > maxx:
                maxx = arr[strt]

            return minn, maxx

        return solve(arr, 0, len(arr) - 1)
```
