```python
Leetcode problem number 239 : https://leetcode.com/problems/sliding-window-maximum/$0
```

```python
class Solution:
    def maxSlidingWindow(self, nums: List[int], k: int) -> List[int]:
        ans = []
        for i in range(len(nums) - k + 1):
            ans.append(max(nums[i:i+k]))
            # ans.append(n)

        return ans
```
