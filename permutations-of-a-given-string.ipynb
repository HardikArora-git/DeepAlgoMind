 GeeksforGreeks problem : https://www.geeksforgeeks.org/problems/permutations-of-a-given-string-1587115620/1

```python
class Solution:
    def permutation(self, s):
        result = []

        def backtrack(s, path):
            if len(s) == 0:
                result.append(path)
                return

            for i in range(len(s)):
                ch = s[i]
                remaining = s[:i] + s[i+1:]
                backtrack(remaining, path + ch)

        backtrack(s, "")
        return result
```
