Leetcode problem number 20 : https://leetcode.com/problems/valid-parentheses/submissions/2039454804/

```python
class Solution:
    def isValid(self, s: str) -> bool:
        stack = []
        mapping = {')':'(','}':'{',']':'['}
        for char in s:
            if char in mapping:
                top = stack.pop() if stack else 'a'
                if mapping[char] != top:
                    return False
            else:
                stack.append(char)
        return len(stack)==0
```
