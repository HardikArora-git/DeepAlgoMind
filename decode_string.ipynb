Leetcode problem number 394 : https://leetcode.com/problems/decode-string/$0  

```python
class Solution:
    def decodeString(self, s: str) -> str:
        stack = []
        num = 0
        newstr = ""

        for st in s:

            if st.isdigit():
                num = num * 10 + int(st)

            elif st == '[':
                stack.append((newstr, num))
                newstr = ""
                num = 0

            elif st == ']':
                prev_str, number  = stack.pop()
                newstr = prev_str + newstr * number
            else:
                newstr += st

        return newstr
```
