leetcode problem number 1021 : https://leetcode.com/problems/remove-outermost-parentheses/submissions/2035321718/

```python
class Solution:
    def removeOuterParentheses(self, s: str) -> str:
        depth = 0
        ans = []
        for c in s:
            if c == '(':
                if depth > 0:
                    ans.append(c)
                depth += 1
            else: # c==')'
                depth -=1
                if depth > 0:
                    ans.append(c)

        return ''.join(ans)
```
